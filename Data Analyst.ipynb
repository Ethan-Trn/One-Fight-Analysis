```python
# Optional: adjust these manually if you already know your desired y-axis range
Y_MIN = 140
Y_MAX = 210

# Load your data
# df = pd.read_csv('your_data.csv')  # <- uncomment and use if loading from CSV

# Clean data: replace invalid height values (e.g., 0) with NaN, then drop them
df['Height (cm)'] = df['Height (cm)'].replace(0, np.nan)
df = df.dropna(subset=['Height (cm)'])

# Optional: adjust y-axis limits based on actual clean data
# Y_MIN = df['Height (cm)'].min()
# Y_MAX = df['Height (cm)'].max()

# Get all unique values
weight_classes = df['Weight Class'].unique()
genders = df['Gender'].unique()

# Loop through each gender and weight class to generate individual boxplots
for gender in genders:
    for wc in weight_classes:
        subset = df[(df['Weight Class'] == wc) & (df['Gender'] == gender)]
        if not subset.empty:
            plt.figure(figsize=(6, 4))
            sns.boxplot(data=subset, x='Weight Class', y='Height (cm)')
            plt.title(f'Boxplot of Height (cm) - {gender} {wc}')
            plt.xlabel('Weight Class')
            plt.ylabel('Height (cm)')
            plt.ylim(Y_MIN, Y_MAX)  # Set consistent y-axis scale
            plt.tight_layout()
            plt.show()
```

```python
# Clean up
df = df.dropna(subset=['Country', 'Gender', 'Sport'])
df['Country'] = df['Country'].str.strip()
df['Country'] = df['Country'].replace({
    'USA': 'United States',
    'U.S.A': 'United States',
    'UK': 'United Kingdom',
    'England': 'United Kingdom',
    'South Korea': 'Korea, South',
    'North Korea': 'Korea, North',
    'Russia': 'Russian Federation'
})

# Get all unique sports
sports = df['Sport'].unique()

# Loop through each sport and plot choropleth
for sport in sports:
    sport_df = df[df['Sport'] == sport]
    country_counts = sport_df['Country'].value_counts().reset_index()
    country_counts.columns = ['Country', 'Fighter Count']
    total = country_counts['Fighter Count'].sum()
    country_counts['Percent of Fighters'] = 100 * country_counts['Fighter Count'] / total

    fig = px.choropleth(
        country_counts,
        locations="Country",
        locationmode="country names",
        color="Percent of Fighters",
        hover_name="Country",
        color_continuous_scale="Reds",
        title=f"{sport}: Percentage of Fighters by Country",
    )
    fig.show()
```

```python
# Ensure there are no missing values in Height or Weight
df_clean = df.dropna(subset=['Height (cm)', 'Weight (kg)', 'Weight Class', 'Gender'])

# Remove "Out of range" weight classes
df_clean = df_clean[df_clean['Weight Class'] != 'Out of range']

# Remove anyone with height less than 100 cm
df_clean = df_clean[df_clean['Height (cm)'] >= 100]

for gender in df_clean['Gender'].unique():
    subset = df_clean[df_clean['Gender'] == gender]
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        data=subset,
        x='Weight (kg)',
        y='Height (cm)',
        hue='Weight Class',
        palette='tab10',
        s=80,
        edgecolor='k'
    )
    plt.title(f'Scatterplot of Weight vs Height by Weight Class ({gender})')
    plt.xlabel('Weight (kg)')
    plt.ylabel('Height (cm)')
    plt.legend(title='Weight Class', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
```

```python
# Rename columns
df = df.rename(columns={
    'WIN': 'Wins',
    'LOSS': 'Losses',
    'Total Fights': 'Total Fight'
})

# Clean and calculate win ratio
df = df.dropna(subset=['Height (cm)', 'Wins', 'Total Fight', 'Weight Class', 'Gender'])
df = df[df['Height (cm)'] >= 100]
df = df[df['Weight Class'] != 'Out of range']
df = df[df['Total Fight'] > 0]
df['Win Ratio'] = df['Wins'] / df['Total Fight']

# Set up a color gradient using seaborn's palette
norm = plt.Normalize(0, 1)  # normalize win ratio from 0 to 1
cmap = sns.color_palette("viridis", as_cmap=True)  # or try "plasma", "coolwarm"

# Plot each (Gender, Weight Class) group
for gender in df['Gender'].unique():
    for wc in df['Weight Class'].unique():
        subset = df[(df['Gender'] == gender) & (df['Weight Class'] == wc)]
        if not subset.empty:
            plt.figure(figsize=(8, 5))
            scatter = plt.scatter(
                subset['Height (cm)'],
                subset['Win Ratio'],
                c=subset['Win Ratio'],
                cmap=cmap,
                norm=norm,
                edgecolor='k',
                alpha=0.8
            )
            plt.colorbar(scatter, label='Win Ratio')
            plt.title(f'Height vs Win Ratio\n{gender} - {wc}')
            plt.xlabel('Height (cm)')
            plt.ylabel('Win Ratio')
            plt.ylim(-0.1, 1.1)
            plt.grid(True, linestyle='--', alpha=0.4)
            plt.tight_layout()
            plt.show()
```

```python
# Make a copy & fill missing method counts with 0
df = df.copy()
df.fillna(0, inplace=True)

# Ensure integer types
win_cols  = ['W TKO', 'W Submission', 'W Unanimous Decision', 'W Split Decision']
loss_cols = ['L TKO', 'L Submission', 'L Unanimous Decision', 'L Split Decision']
df[win_cols + loss_cols] = df[win_cols + loss_cols].astype(int)

def determine_style(row):
    tko = row['W TKO']
    sub = row['W Submission']
    dec = row['W Unanimous Decision'] + row['W Split Decision']

    # Well-Rounded if both striking & submission are >1 and within 2 wins of each other
    if tko > 1 and sub > 1 and abs(tko - sub) <= 2:
        return 'Well Rounded'
    # Otherwise pick the dominant method
    if tko > sub and tko > dec:
        return 'Striker'
    if sub > tko and sub > dec:
        return 'Wrestler'
    if dec > tko and dec > sub:
        return 'Decision Machine'
    return 'Unclear'

def determine_counter_style(row):
    ltko = row['L TKO']
    lsub = row['L Submission']
    ldec = row['L Unanimous Decision'] + row['L Split Decision']

    if ltko > lsub and ltko > ldec:
        return 'Striker'
    if lsub > ltko and lsub > ldec:
        return 'Wrestler'
    if ldec > ltko and ldec > lsub:
        return 'Decision Machine'
    return 'Unclear'

# Apply and create new columns
df['Style'] = df.apply(determine_style, axis=1)
df['Counter Style'] = df.apply(determine_counter_style, axis=1)

# Now df has two new columns you can inspect or export
# Example: df.to_csv("styled_athletes.csv", index=False)
```
