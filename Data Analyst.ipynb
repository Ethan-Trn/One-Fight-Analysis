```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
import os
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore
df = pd.read_csv('C:\\Users\\ethan\\Downloads\\One-Fight-Analysis\\athlete_stats.csv')
df.head(10)
```

```python
#Findout the most optimal Height (cm) (cm) (cm) (cm) (cm) (cm) (cm) for each weight class
#Create a new column called weight_class into df 
#The weight class is determined by the following ranges for Males:
    #105lbs to 115lbs = AtomWeight
    #115lbs to 125lbs = StrawWeight
    #125lbs to 135lbs = FlyWeight
    #135lbs to 145lbs = BantamWeight
    #145lbs to 155lbs = FeatherWeight
    #155lbs to 170lbs = Lightweight
    #170lbs to 185lbs = WelterWeight
    #185lbs to 205lbs = MiddleWeight
    #205lbs to 225lbs = Light HeavyWeight
    #225lbs to 265lbs = HeavyWeight
bins = [105, 115, 125, 135, 145, 155, 170, 185, 205, 225, 265]
bins = [x * 0.453592 for x in bins] # Convert lbs to kg
df['Weight (kg)'] = df['Weight (kg)'].replace(-1, np.nan)
labels = ['AtomWeight', 'StrawWeight', 'FlyWeight', 'BantamWeight', 'FeatherWeight', 'Lightweight', 'WelterWeight', 'MiddleWeight', 'Light HeavyWeight', 'HeavyWeight']
df['weight_class'] = pd.cut(df['Weight (kg)'], bins=bins, labels=labels, right=False)
# print (df['Weight (kg)'].dtype)
df['total_fights'] = df['WIN'] + df['LOSS'] + df['DRAW']
df['Win Rate'] = df['WIN'] / df['total_fights']
# print(df[['Name','weight_class','Win Rate', 'WIN','total_fights']].head(10))
# Filter the DataFrame for rows where 'Weight (kg)' is -1 and 'total_fights' > 1
# Filter the DataFrame for rows where 'Weight (kg)' is NaN and 'total_fights' > 1
filtered_df = df[df['Weight (kg)'].isna() & (df['total_fights'] > 1)]

# Print the filtered DataFrame
print(filtered_df[['Name','total_fights']])
```
