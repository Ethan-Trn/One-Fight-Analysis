```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
import os
import matplotlib.pyplot as plt # type: ignore
import seaborn as sns # type: ignore

df = pd.read_csv('C:\\Users\\ethan\\Downloads\\One-Fight-Analysis\\athlete_stats.csv')
df.head(10)
```

```python
#Findout the most optimal Height (cm) (cm) (cm) (cm) (cm) (cm) (cm) for each weight class
#Create a new column called weight_class into df 
#The weight class is determined by the following ranges for Males:
    #105lbs to 115lbs = AtomWeight
    #115lbs to 125lbs = StrawWeight
    #125lbs to 135lbs = FlyWeight
    #135lbs to 145lbs = BantamWeight
    #145lbs to 155lbs = FeatherWeight
    #155lbs to 170lbs = Lightweight
    #170lbs to 185lbs = WelterWeight
    #185lbs to 205lbs = MiddleWeight
    #205lbs to 225lbs = Light HeavyWeight
    #225lbs to 265lbs = HeavyWeight
bins = [105, 115, 125, 135, 145, 155, 170, 185, 205, 225, 265]
labels = ['AtomWeight', 'StrawWeight', 'FlyWeight', 'BantamWeight', 'FeatherWeight', 'Lightweight', 'WelterWeight', 'MiddleWeight', 'Light HeavyWeight', 'HeavyWeight']
df['weight_class'] = pd.cut(df['Weight (kg)'], bins=bins, labels=labels, right=False)
df['Win Rate'] = df['WIN'] / df['WIN'] + df['LOSS'] + df['DRAW']
```

```python
#Make it so that each height is rounded to the nearest inch
df['Height (cm) (cm)'] = df['Height (cm) (cm)'] / 2.54
df['Height (cm)'] = df['Height (cm)'].round(0)

#Then Calculate the average amount of win for each weight class and height
#Definiely not efficent but for now it works
#First create df based on the weightclass and Gender
# Male weight classes
M_Atomweight = df[(df['weight_class'] == 'AtomWeight') & (df['Gender'] == 'Male')]
M_Strawweight = df[(df['weight_class'] == 'StrawWeight') & (df['Gender'] == 'Male')]
M_Flyweight = df[(df['weight_class'] == 'FlyWeight') & (df['Gender'] == 'Male')]
M_Bantamweight = df[(df['weight_class'] == 'BantamWeight') & (df['Gender'] == 'Male')]
M_Featherweight = df[(df['weight_class'] == 'FeatherWeight') & (df['Gender'] == 'Male')]
M_Lightweight = df[(df['weight_class'] == 'Lightweight') & (df['Gender'] == 'Male')]
M_Welterweight = df[(df['weight_class'] == 'WelterWeight') & (df['Gender'] == 'Male')]
M_Middleweight = df[(df['weight_class'] == 'MiddleWeight') & (df['Gender'] == 'Male')]
M_LightHeavyweight = df[(df['weight_class'] == 'Light HeavyWeight') & (df['Gender'] == 'Male')]
M_Heavyweight = df[(df['weight_class'] == 'HeavyWeight') & (df['Gender'] == 'Male')]

# Female weight classes (stop at Bantamweight)
F_Atomweight = df[(df['weight_class'] == 'AtomWeight') & (df['Gender'] == 'Female')]
F_Strawweight = df[(df['weight_class'] == 'StrawWeight') & (df['Gender'] == 'Female')]
F_Flyweight = df[(df['weight_class'] == 'FlyWeight') & (df['Gender'] == 'Female')]
F_Bantamweight = df[(df['weight_class'] == 'BantamWeight') & (df['Gender'] == 'Female')]

```

```python
#Then For Each Height Within The Weight Class Get The Average Win Rate
#Like 65 inches has a win rate of 0.5 and 66 inches has a win rate of 0.6
#Then Get The Max Win Rate For Each Weight Class
#Make it that only people with more than 2 fights are considered

#This is the one that filters
#Atomweight = Atomweight[Atomweight['WIN'] + Atomweight['LOSS'] + Atomweight['DRAW'] > 2]
#This is the one that groups by height and gets the average win rate
#Atomweight = Atomweight.groupby('height')['Win Rate'].mean().reset_index()

#male
M_Atomweight = M_Atomweight[M_Atomweight['WIN'] + M_Atomweight['LOSS'] + M_Atomweight['DRAW'] > 2]
M_Atomweight = M_Atomweight.groupby('height')['Win Rate'].mean().reset_index()
M_Atomweight['weight_class'] = 'AtomWeight'

M_Strawweight = M_Strawweight[M_Strawweight['WIN'] + M_Strawweight['LOSS'] + M_Strawweight['DRAW'] > 2]
M_Strawweight = M_Strawweight.groupby('height')['Win Rate'].mean().reset_index()
M_Strawweight['weight_class'] = 'StrawWeight'

M_Flyweight = M_Flyweight[M_Flyweight['WIN'] + M_Flyweight['LOSS'] + M_Flyweight['DRAW'] > 2]
M_Flyweight = M_Flyweight.groupby('height')['Win Rate'].mean().reset_index()
M_Flyweight['weight_class'] = 'FlyWeight'

M_Bantamweight = M_Bantamweight[M_Bantamweight['WIN'] + M_Bantamweight['LOSS'] + M_Bantamweight['DRAW'] > 2]
M_Bantamweight = M_Bantamweight.groupby('height')['Win Rate'].mean().reset_index()
M_Bantamweight['weight_class'] = 'BantamWeight'

M_Featherweight = M_Featherweight[M_Featherweight['WIN'] + M_Featherweight['LOSS'] + M_Featherweight['DRAW'] > 2]
M_Featherweight = M_Featherweight.groupby('height')['Win Rate'].mean().reset_index()
M_Featherweight['weight_class'] = 'FeatherWeight'

M_Lightweight = M_Lightweight[M_Lightweight['WIN'] + M_Lightweight['LOSS'] + M_Lightweight['DRAW'] > 2]
M_Lightweight = M_Lightweight.groupby('height')['Win Rate'].mean().reset_index()
M_Lightweight['weight_class'] = 'Lightweight'

M_Welterweight = M_Welterweight[M_Welterweight['WIN'] + M_Welterweight['LOSS'] + M_Welterweight['DRAW'] > 2]
M_Welterweight = M_Welterweight.groupby('height')['Win Rate'].mean().reset_index()
M_Welterweight['weight_class'] = 'WelterWeight'

M_Middleweight = M_Middleweight[M_Middleweight['WIN'] + M_Middleweight['LOSS'] + M_Middleweight['DRAW'] > 2]
M_Middleweight = M_Middleweight.groupby('height')['Win Rate'].mean().reset_index()
M_Middleweight['weight_class'] = 'MiddleWeight'

M_LightHeavyweight = M_LightHeavyweight[M_LightHeavyweight['WIN'] + M_LightHeavyweight['LOSS'] + M_LightHeavyweight['DRAW'] > 2]
M_LightHeavyweight = M_LightHeavyweight.groupby('height')['Win Rate'].mean().reset_index()
M_LightHeavyweight['weight_class'] = 'Light HeavyWeight'

M_Heavyweight = M_Heavyweight[M_Heavyweight['WIN'] + M_Heavyweight['LOSS'] + M_Heavyweight['DRAW'] > 2]
M_Heavyweight = M_Heavyweight.groupby('height')['Win Rate'].mean().reset_index()
M_Heavyweight['weight_class'] = 'HeavyWeight'

# Female weight classes (stop at Bantamweight)
F_Atomweight = F_Atomweight[F_Atomweight['WIN'] + F_Atomweight['LOSS'] + F_Atomweight['DRAW'] > 2]
F_Atomweight = F_Atomweight.groupby('height')['Win Rate'].mean().reset_index()
F_Atomweight['weight_class'] = 'AtomWeight'

F_Strawweight = F_Strawweight[F_Strawweight['WIN'] + F_Strawweight['LOSS'] + F_Strawweight['DRAW'] > 2]
F_Strawweight = F_Strawweight.groupby('height')['Win Rate'].mean().reset_index()
F_Strawweight['weight_class'] = 'StrawWeight'

F_Flyweight = F_Flyweight[F_Flyweight['WIN'] + F_Flyweight['LOSS'] + F_Flyweight['DRAW'] > 2]
F_Flyweight = F_Flyweight.groupby('height')['Win Rate'].mean().reset_index()
F_Flyweight['weight_class'] = 'FlyWeight'

F_Bantamweight = F_Bantamweight[F_Bantamweight['WIN'] + F_Bantamweight['LOSS'] + F_Bantamweight['DRAW'] > 2]
F_Bantamweight = F_Bantamweight.groupby('height')['Win Rate'].mean().reset_index()
F_Bantamweight['weight_class'] = 'BantamWeight'
```

```python
# Combine all weight classes into a single DataFrame for visualization
all_classes = pd.concat([
    M_Atomweight, M_Strawweight, M_Flyweight, M_Bantamweight, M_Featherweight,
    M_Lightweight, M_Welterweight, M_Middleweight, M_LightHeavyweight, M_Heavyweight,
    F_Atomweight, F_Strawweight, F_Flyweight, F_Bantamweight
])

# Bar Graph: Win Rate by Height for Each Weight Class and Gender
plt.figure(figsize=(15, 8))
sns.barplot(data=all_classes, x='height', y='Win Rate', hue='weight_class', dodge=True)
plt.title('Win Rate by Height for Each Weight Class and Gender')
plt.xlabel('Height (inches)')
plt.ylabel('Win Rate')
plt.legend(title='Weight Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Scatter Plot: Win Rate by Height for Each Weight Class and Gender
plt.figure(figsize=(15, 8))
sns.scatterplot(data=all_classes, x='height', y='Win Rate', hue='weight_class', style='weight_class', s=100)
plt.title('Scatter Plot of Win Rate by Height for Each Weight Class and Gender')
plt.xlabel('Height (inches)')
plt.ylabel('Win Rate')
plt.legend(title='Weight Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
